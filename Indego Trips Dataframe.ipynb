```python
import pandas as pd

indego_trips_2024_q3 = pd.read_csv('/content/drive/MyDrive/Indego Project/indego-trips-2024-q3.csv')
indego_trips_2024_q4 = pd.read_csv('/content/drive/MyDrive/Indego Project/indego-trips-2024-q4.csv')
indego_trips_2025_q1 = pd.read_csv('/content/drive/MyDrive/Indego Project/indego-trips-2025-q1.csv')
indego_trips_2025_q2 = pd.read_csv('/content/drive/MyDrive/Indego Project/indego-trips-2025-q2.csv')
indego_trips_2025_q3 = pd.read_csv('/content/drive/MyDrive/Indego Project/indego-trips-2025-q3.csv')
```

```python
indego_df = pd.concat([indego_trips_2024_q3, indego_trips_2024_q4, indego_trips_2025_q1, indego_trips_2025_q2, indego_trips_2025_q3], ignore_index=True)
```

```python
indego_df.head()
```

```python
indego_df['date_column'] = pd.to_datetime(indego_df['start_time'])
```

```python
indego_df = indego_df[(indego_df['date_column'] >= '2024-08-01') & (indego_df['date_column'] <= '2025-08-25')]
```

```python
print(indego_df['date_column'].min())  # should be 2024-08-01
print(indego_df['date_column'].max())  # should be 2025-08-01
```

```python
indego_df.to_csv('Indego_trips_2024_2025.csv', index=False)
```
